```python
import os
import random
import shutil

data_path = "./unprep_data/images"

# path to destination folders
train_folder = "./data/images/train"
val_folder = "./data/images/val"

# Define a list of image extensions
image_extensions = ['.jpg', '.jpeg', '.png', '.bmp']

# Create a list of image filenames in 'data_path'
imgs_list = [filename for filename in os.listdir(data_path) if os.path.splitext(filename)[-1] in image_extensions]

# Sets the random seed 
random.seed(42)

# Shuffle the list of image filenames
random.shuffle(imgs_list)

# determine the number of images for each set
train_size = 457
val_size = 115

# Create destination folders if they don't exist
for folder_path in [train_folder, val_folder]:
    if not os.path.exists(folder_path):
        os.makedirs(folder_path)

# Copy image files to destination folders
for i, f in enumerate(imgs_list):
    if i < train_size:
        dest_folder = train_folder
    elif i < train_size + val_size:
        dest_folder = val_folder
    
    shutil.copy(os.path.join(data_path, f), os.path.join(dest_folder, f))
```

```python
for folder in ['train','val']:
    for file in os.listdir(os.path.join('data', 'images', folder)):
        
        filename = file.split('.')[0]+'.txt'
        existing_filepath = os.path.join('unprep_data','labels', filename)
        if os.path.exists(existing_filepath): 
            new_filepath = os.path.join('data','labels', folder, filename)
            os.replace(existing_filepath, new_filepath)    
```

```python


from ultralytics import YOLO
import cv2


model_path = os.path.join( 'runs', 'detect', 'train', 'weights', 'best.pt')
model = YOLO(model_path)

video_path = os.path.join('fire.mp4')
video_path_out = '{}_out.mp4'.format(video_path)

cap = cv2.VideoCapture(video_path)
ret, frame = cap.read()
H, W, _ = frame.shape
out = cv2.VideoWriter(video_path_out, cv2.VideoWriter_fourcc(*'MP4V'), int(cap.get(cv2.CAP_PROP_FPS)), (W, H))


# Load a model
  # load a custom model

threshold = 0.5

while ret:

    results = model(frame)[0]
    #print(len(results))
    for result in results.boxes.data.tolist():
        x1, y1, x2, y2, score, class_id = result

        if score > threshold:
            cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 4)
            cv2.putText(frame, results.names[int(class_id)].upper(), (int(x1), int(y1 - 10)),
                        cv2.FONT_HERSHEY_SIMPLEX, 1.3, (0, 255, 0), 3, cv2.LINE_AA)

    out.write(frame)
    ret, frame = cap.read()

cap.release()
out.release()
cv2.destroyWindow(cap)
cv2.destroyWindow(out)
```
